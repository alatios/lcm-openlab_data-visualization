# Imshow

```python
cmat = df[['L_ENDVERTEX_X', 'L_ENDVERTEX_Y', 'L_ENDVERTEX_Z', 'L_PX', 'L_PY', 'L_PZ']].corr()
```

```python
cmat
```

```python
fig, ax = plt.subplots()

ims = ax.imshow(cmat, cmap='Spectral_r')
ims.set_clim(-0.2,0.2)

ax.set_xticks(np.arange(6))
ax.set_xticklabels(cmat.columns, rotation=90)

ax.set_yticks(np.arange(6))
ax.set_yticklabels(cmat.columns)

for i in range(len(cmat)):
    for j in range(len(cmat)):
        if i==j:
            color = 'w'
        else:
            color = 'k'
                   
        ax.text(i, j, "{0:.2f}".format(cmat.iloc[i,j]), ha='center', va='center', color=color)

fig.colorbar(ims, label='Correlation')

plt.show()
```

# Plotting a fit in Python

```python
np.random.seed(2022)
```

```python
data = np.random.normal(0, 1, size=int(1e4))
```

```python
n, bins, _ = plt.hist(data, bins=80)
plt.close()
```

```python
len(n)
```

```python
len(bins)
```

```python
midpoints = (bins[1:] + bins[:-1])/2
```

```python
def gaussian(x, mu, sigma):
    return 1/np.sqrt(2*np.pi) * np.exp(-0.5 * (x-mu)**2 / sigma**2)
```

```python
fig, axs = plt.subplots(2, 1, sharex=True, figsize=(5,6), gridspec_kw={'height_ratios': [3, 1]})
fig.subplots_adjust(hspace=0)

ax = axs[0]

ax.errorbar(midpoints, n, np.sqrt(n), linestyle='', marker='.', ecolor='k', capsize=2, label='Data')
ax.plot(midpoints, 1000*gaussian(midpoints,0, 1), label='Fit')

ax.legend(frameon=False)
ax.set_ylabel('Events')

ax = axs[1]

ax.errorbar(midpoints, (n - 1000*gaussian(midpoints, 0, 1))/np.sqrt(n), 1., linestyle='', marker='.', ecolor='k')
ax.set_ylim(-3, 3)
ax.grid(linestyle='--')
ax.axhline(0, color='tab:orange')

plt.show()
```

Check out `scipy.optimize.curve_fit()` to do an actual binned fit in Python!

```python
fig, axs = plt.subplots(2, 1, sharex=True, figsize=(5,6), gridspec_kw={'height_ratios': [3, 1]})
fig.subplots_adjust(hspace=0)

ax = axs[0]

ax.errorbar(midpoints, n, np.sqrt(n), linestyle='', marker='.', ecolor='k', capsize=2, label='Data')
ax.plot(midpoints, 1000*gaussian(midpoints,0, 1), label='Fit')

ax.legend(frameon=False)
ax.set_ylabel('Events')

ax = axs[1]

ax.bar(x=midpoints, height=(n - 1000*gaussian(midpoints, 0, 1))/np.sqrt(n), width=.04, color='tab:gray')
ax.set_ylim(-3, 3)
ax.grid(linestyle='--')

plt.show()
```
